```python
def regex_depuration(string):
    re_pattern_dict = {'caps':r'\n\n\b[A-ZÁÉÍÓÚÑÜ]+-[A-ZÁÉÍÓÚÑÜ]+\b\s\n\n', 'page_nums': r'\d+\s\n\n\n'}
    for r in re_pattern_dict.values():
        pattern = re.compile(r)
        # para aplicar los cambios del reemplazo del patrón anterior, es necesario actualizar el string en cada iteración para que se conserven los cambios.
        string = pattern.sub('', string)
    
    return string

regex_depuration(test_pulpo)
```

```python
def molinizador(text):
    # para facilitar la legibilidad del json, quitaremos saltos de línea del txt y guiones.
    to_remove = ['\n', '- ', '- \n']
    test_cap_split = [w for w in text.split("\n\n\n")]

    words = []
    definitions = []
    for t in test_cap_split:
        if t != "" and t.split(" ")[0].islower():
            words.append(t.split(" ")[0].replace(",", ""))
            definition = " ".join(t.split(" ")[1::])
            for t in to_remove:
                definition = definition.replace(t, "").strip()
            
            definitions.append(definition)

    mm_dict = {}
    for w, d in zip(words, definitions):
        mm_dict[w] = d

    return mm_dict
```
